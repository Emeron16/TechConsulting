```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import plotly.figure_factory as ff
```

# Exploratory data Analaysis

```python
df = pd.read_csv("datasets/diabetes.csv")
```

```python
df.info()
```

```python
df.head()
```

# Data Cleaning

```python
# change the , to periods for the columns and also make the col float type instead of string
cols = ['chol_hdl_ratio', 'bmi', 'waist_hip_ratio']
for col in cols:
    df[col] = df[col].str.replace(',', '.')
    df[col] = pd.to_numeric(df[col])

df
```

```python
df.info()
```

```python
# use one hote encoding to convert the gender, and diabetes column to bools
df = pd.get_dummies(df, columns=[['gender','diabetes']], drop_first=True)
```

```python
df
df.info()
```

# Training

```python
X = df.drop(columns=['diabetes_No diabetes'])
y = df['diabetes_No diabetes']
```

```python
# split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=2)
```

# Decision Tree

```python
# Create and train 
model=DecisionTreeClassifier(max_depth=5, random_state=42)
model.fit(X_train, y_train)
```

```python
# Predict
y_pred = model.predict(X_test)
y_pred
```

```python
# Recall — "Of all actual cases of this class, how many did the model catch?"
# Precision — "Of all the times the model predicted this class, how often was it right?"
#F1-Score — Balanced average of precision and recall. Use this when the classes are imbalanced.
#Support — How many actual samples of each class are in the test set
print(classification_report(y_test, y_pred))
```

```python
dc_cm = confusion_matrix(y_test, y_pred)

labels = ['No diabetes', 'Diabetes']

fig = ff.create_annotated_heatmap(
    z=dc_cm,
    x=labels,
    y=labels,
    colorscale='Blues',
    showscale=True
)
fig.update_layout(
    title='Confusion Matrix',
    xaxis_title='Predicted',
    yaxis_title='Actual'
)
fig.show()
```
