# Data Cleaning

```python
# Drop unneeded columns
df.drop(columns=['patient_number'], inplace=True)
```

```python
# change the , to periods for the columns and also make the col float type instead of string
cols = ['chol_hdl_ratio', 'bmi', 'waist_hip_ratio']
for col in cols:
    df[col] = df[col].str.replace(',', '.')
    df[col] = pd.to_numeric(df[col])

df
```

```python
df.info()
```

```python
# use one hote encoding to convert the gender, and diabetes column to bools
df = pd.get_dummies(df, columns=['gender','diabetes'], drop_first=True)
```

# Training

```python
X = df.drop(columns=['diabetes_No diabetes'])
y = df['diabetes_No diabetes']
```

```python
# split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=2)
```

# Decision Tree

```python
# Create and train 
model=DecisionTreeClassifier(max_depth=5, random_state=42)
model.fit(X_train, y_train)
```

```python
# Predict
y_pred = model.predict(X_test)
y_pred
```

```python
dt_y_pred = model.predict(X_train)
print(classification_report(y_train, dt_y_pred))
```

```python
# Recall — "Of all actual cases of this class, how many did the model catch?"
# Precision — "Of all the times the model predicted this class, how often was it right?"
#F1-Score — Balanced average of precision and recall. Use this when the classes are imbalanced.
#Support — How many actual samples of each class are in the test set
print(classification_report(y_test, y_pred))
```

```python
dc_cm = confusion_matrix(y_test, y_pred)

labels = ['No diabetes', 'Diabetes']

fig = ff.create_annotated_heatmap(
    z=dc_cm,
    x=labels,
    y=labels,
    colorscale='Blues',
    showscale=True
)
fig.update_layout(
    title='Confusion Matrix',
    xaxis_title='Predicted',
    yaxis_title='Actual'
)
fig.show()
```

```python
importance_df = pd.DataFrame({
    'feature': X_train.columns,
    'importance': model.feature_importances_
}).sort_values('importance', ascending=False)

fig = px.bar(
    importance_df,
    x='feature',
    y='importance',
    title='Decision Feature Importances'
)
fig.show()
```

# Random Forest

```python
#n_estimators: number of base learners
rfc = RandomForestClassifier(n_estimators=100, max_depth=25, random_state=2) 
rfc.fit(X_train, y_train)
```

```python
rfc_y_pred = rfc.predict(X_test)
```

```python
# Performance on test data
print(classification_report(y_test, rfc_y_pred))
```

```python
# Performance on training data
rfc_y_train_pred = rfc.predict(X_train)
print(classification_report(y_train, rfc_y_train_pred))
```

```python

cm = confusion_matrix(y_test, rfc_y_pred)

labels = ['No Diabetes', 'Diabetes']

fig = ff.create_annotated_heatmap(
    z=cm,
    x=labels,
    y=labels,
    colorscale='Blues',
    showscale=True
)

fig.update_layout(
    title='Confusion Matrix',
    xaxis_title='Predicted',
    yaxis_title='Actual'
)

fig.show()
```

```python
# View which feature affected the models decision the most
importance_df = pd.DataFrame({
    'feature': X_train.columns,
    'importance': rfc.feature_importances_
}).sort_values('importance', ascending=False)

fig = px.bar(
    importance_df,
    x='feature',
    y='importance',
    title='Random Forest Feature Importances'
)
fig.show()
```
